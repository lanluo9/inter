```python
import os
import pandas as pd

dir_mat_inter = r'\\duhs-user-nc1.dhe.duke.edu\dusom_glickfeldlab\All_Staff\home\lan\Data\2P_images\mat_inter/'.replace('\\', '/')
dir_file = os.path.join(dir_mat_inter, 'adp_dataset_master.xlsx')
data_info = pd.read_excel(dir_file)

metadata = data_info[(~data_info.paradigm.str.contains('lindsey'))
                     & (data_info.paradigm != 'grating_2ori_multisess')
                    & ((data_info.cellpose_seg == True) | (data_info.manual_seg == True))
                    ].reset_index(drop=True)
```

```python
df_cold = metadata.groupby(['mouse', 'date']).first().reset_index() # these mouse-date subdir go to cold storage
df_data = r'Z:\All_Staff\home\lan\Data\2P_images/'.replace('\\', '/')

for irow in range(df_cold.shape[0]):
    mouse = df_cold.loc[irow, 'mouse'].astype(str)
    date = df_cold.loc[irow, 'date'].astype(str)
    dir_mouse_date = os.path.join(df_data, mouse, date, '').replace('\\', '/')
    print(dir_mouse_date)
    print('')
```
